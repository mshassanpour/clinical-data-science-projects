# Exploratory Data Analysis

Four tables will be merged into one table "combined_patient_data"


```python

#merging the four tables as a new table

db_path = '/content/data_split/patient_data.db'

# Connect to the SQLite database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Step 1: Create a new table to store the combined data
# Adjust the column definitions
try:
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS combined_patient_data (
        patient_id TEXT PRIMARY KEY,
            diagnosis_code TEXT,
            totChol REAL,
            sysBP  REAL,
            diaBP  REAL,
            BMI    REAL,
            heartRate  REAL,
            glucose    REAL,
            TenYearCHD INTEGER,
            male	INTEGER,
            age	INTEGER,
            education INTEGER,
            currentSmoker INTEGER,
             cigsPerDay INTEGER,
            BPMeds INTEGER,
            prevalentStroke INTEGER,
            prevalentHyp INTEGER,
            diabetes INTEGER,
            Note TEXT
    )
    ''')
    print("Table 'combined_patient_data' created successfully.")
except sqlite3.Error as e:
    print(f"An error occurred while creating the table: {e}")

# Step 2: Define the SQL query for joining the four tables
sql_join_query =  '''
SELECT
    pc.patient_id, pc.diagnosis_code,
    pc.totChol, pc.sysBP, pc.diaBP,BMI,pc.heartRate,pc.glucose,pc.TenYearCHD,
    pd.male,pd.age, pd.education,
    ph.currentSmoker, ph.cigsPerDay,	ph.BPMeds,	ph.prevalentStroke,
    ph.prevalentHyp,	ph.diabetes,
    pn.Note
FROM
    patient_clinical_table pc
LEFT JOIN
    patient_demographics_table pd ON pc.patient_id = pd.patient_id
LEFT JOIN
    patient_history_table ph ON pc.patient_id = ph.patient_id
LEFT JOIN
    patient_note_table pn ON pc.patient_id = pn.patient_id
'''

# Step 3: Insert the joined data into the new table
try:
    # Execute the join query
    cursor.execute(sql_join_query)
    joined_data = cursor.fetchall()

    # Insert data into the new table
    insert_query = '''
    INSERT INTO combined_patient_data
    (patient_id, diagnosis_code,totChol,sysBP,diaBP, BMI,
            heartRate,glucose,TenYearCHD,male,age,education,
            currentSmoker,cigsPerDay,BPMeds,prevalentStroke,
            prevalentHyp, diabetes,Note)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?,?,?,?,?,?,?,?,?,?,?,?)
    '''
    cursor.executemany(insert_query, joined_data)
    conn.commit()
    print(f"Successfully inserted {cursor.rowcount} rows into 'combined_patient_data'.")
except sqlite3.Error as e:
    print(f"An error occurred: {e}")
finally:
    # Close the connection to the database
    conn.close()
```

Column statistics

```python


# Path to  SQLite database
db_path = '/content/data_split/patient_data.db'
table_name = 'combined_patient_data'  # Specify  table name here

# Connect to the SQLite database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

def get_numerical_columns(cursor, table_name):
    """
    Retrieve the names of numerical columns in a given table.
    """
    cursor.execute(f'PRAGMA table_info({table_name});')
    columns_info = cursor.fetchall()
    numerical_columns = [column[1] for column in columns_info if column[2] in ('REAL', 'INTEGER')]
    return numerical_columns

def calculate_statistics(cursor, table_name, numerical_columns):
    """
    Calculate and print the mean, minimum, maximum, and standard deviation
    of numerical columns in the specified table.
    """
    for column_name in numerical_columns:
        cursor.execute(f'''
            SELECT
                AVG({column_name}),
                MIN({column_name}),
                MAX({column_name}),
                ROUND(
                    SQRT(
                        AVG({column_name} * {column_name})
                        - AVG({column_name}) * AVG({column_name})
                    ),
                2)
            FROM {table_name}
        ''')
        mean, minimum, maximum, std_dev = cursor.fetchone()
        print(f"Statistics for '{column_name}':")
        print(f"  Mean: {mean}")
        print(f"  Minimum: {minimum}")
        print(f"  Maximum: {maximum}")
        print(f"  Standard Deviation: {std_dev}")

try:
    # Get the names of numerical columns in the specified table
    numerical_columns = get_numerical_columns(cursor, table_name)

    # Calculate and print the statistics for numerical columns
    calculate_statistics(cursor, table_name, numerical_columns)

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection to the database
    conn.close()
```

```python
# Path to SQLite database
db_path = '/content/data_split/patient_data.db'
table_name = 'combined_patient_data'

# Connect to the SQLite database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

def get_column_names(cursor, table_name):
    """
    Retrieve the names of columns in a given table.
    """
    cursor.execute(f'PRAGMA table_info({table_name});')
    columns_info = cursor.fetchall()
    column_names = [column[1] for column in columns_info]
    return column_names

def generate_tables_by_grouping(cursor, table_name, column_names):
    """
    Generate tables by performing GROUP BY operation on columns with less than 15 unique values.
    """
    for column_name in column_names:
        cursor.execute(f'SELECT {column_name}, COUNT(*) FROM {table_name} GROUP BY {column_name}')
        group_data = cursor.fetchall()

        num_unique_values = len(group_data)

        if num_unique_values <= 15:
            print(f"Generating table based on '{column_name}'...")
            try:
                cursor.execute(f'DROP TABLE IF EXISTS {column_name}_grouped;')
                cursor.execute(f'CREATE TABLE {column_name}_grouped ({column_name} TEXT, count INTEGER);')
                cursor.executemany(f'INSERT INTO {column_name}_grouped VALUES (?, ?);', group_data)
                conn.commit()
                print(f"Table '{column_name}_grouped' generated successfully.")
            except sqlite3.Error as e:
                print(f"An error occurred while generating table for '{column_name}': {e}")

try:
    # Get the names of columns in the specified table
    column_names = get_column_names(cursor, table_name)

    # Generate tables by grouping data based on columns with less than 15 unique values
    generate_tables_by_grouping(cursor, table_name, column_names)

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection to the database
    conn.close()
```

```python


# Path to  SQLite database
db_path = '/content/data_split/patient_data.db'
table_name = 'combined_patient_data'  # Specify  table name here

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

try:
    # Read the data from the SQLite database into a pandas DataFrame
    df = pd.read_sql_query(f'SELECT * FROM {table_name}', conn)

    # Drop non-numeric columns
    numerical_columns = df.select_dtypes(include=['number']).columns

    # Find outliers in each numerical column
    outliers = {}
    for col in numerical_columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outliers[col] = col_outliers.tolist()

    # Print the outliers for each numerical column
    for col, col_outliers in outliers.items():
        print(f"Outliers in '{col}':")
        print(col_outliers)
        print()

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection to the database
    conn.close()
```

```python
import numpy as np
from scipy.stats import skew, kurtosis, norm


normality_threshold = 2.0  # Threshold for skewness and kurtosis

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

try:
    # Read the data from the SQLite database into a pandas DataFrame
    df = pd.read_sql_query(f'SELECT * FROM {table_name}', conn)

    # Identify numerical columns with over 10 unique values
    numerical_columns = df.select_dtypes(include=['number']).columns
    columns_to_check = [col for col in numerical_columns if df[col].nunique() > 10]

    # Check distribution of each numerical column
    for col in columns_to_check:
        skewness = skew(df[col])
        kurt = kurtosis(df[col])

        # Check if skewness and kurtosis are within the threshold
        if abs(skewness) < normality_threshold and abs(kurt - 3) < normality_threshold:
            print(f"Column '{col}' has a distribution close to normal.")
            print(f"Skewness: {skewness:.2f}, Kurtosis: {kurt:.2f}")

            # Plot histogram and normal distribution curve for visualization
            plt.figure(figsize=(8, 6))
            sns.histplot(data=df, x=col, bins=20, kde=True, stat='density', color='blue', label='Histogram')
            xmin, xmax = plt.xlim()
            x = np.linspace(xmin, xmax, 100)
            p = norm.pdf(x, df[col].mean(), df[col].std())
            plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
            plt.title(f'Distribution of {col} with Normal Distribution Curve')
            plt.xlabel(col)
            plt.ylabel('Density')
            plt.legend()
            plt.show()
            print()  # Add a blank line between plots

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection to the database
    conn.close()
#these two distribution are more in agreement with normal dist
```

```python


# Connect to the SQLite database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

try:
    # Define SQL query to calculate quartiles for 'cigsPerDay'
    quartiles_query = f"""
    SELECT
        MIN(cigsPerDay) AS min_cigs,
        (SELECT cigsPerDay FROM {table_name} ORDER BY cigsPerDay LIMIT 1 OFFSET (SELECT COUNT(*) FROM {table_name}) / 4) AS q1_cigs,
        (SELECT cigsPerDay FROM {table_name} ORDER BY cigsPerDay LIMIT 1 OFFSET (SELECT COUNT(*) FROM {table_name}) * 2 / 4) AS q2_cigs,
        (SELECT cigsPerDay FROM {table_name} ORDER BY cigsPerDay LIMIT 1 OFFSET (SELECT COUNT(*) FROM {table_name}) * 3 / 4) AS q3_cigs,
        MAX(cigsPerDay) AS max_cigs
    FROM
        {table_name}
    """

    # Execute the query to calculate quartiles
    cursor.execute(quartiles_query)
    quartiles_result = cursor.fetchone()

    min_cigs, q1_cigs, q2_cigs, q3_cigs, max_cigs = quartiles_result

    # Define thresholds for categorizing 'cigsPerDay' into 'many', 'average', and 'few' groups
    many_threshold = q3_cigs + (1.5 * (q3_cigs - q1_cigs))
    average_threshold = q1_cigs + (0.5 * (q3_cigs - q1_cigs))

    # Print thresholds for each group
    print(f"Thresholds for 'many' category:")
    print(f"- Greater than or equal to: {many_threshold}")
    print(f"Thresholds for 'average' category:")
    print(f"- Greater than or equal to: {average_threshold}")
    print(f"Thresholds for 'few' category:")
    print(f"- Less than: {average_threshold}")
    print()

    # Define SQL query to categorize 'cigsPerDay' into 'many', 'average', and 'few' groups
    query = f"""
    SELECT
        CASE
            WHEN cigsPerDay >= {many_threshold} THEN 'many'
            WHEN cigsPerDay >= {average_threshold} THEN 'average'
            ELSE 'few'
        END AS cigs_category,
        AVG(age) AS avg_age,
        AVG(heartRate) AS avg_heart_rate,
        AVG(sysBP) AS avg_sysBP,
        AVG(glucose) AS avg_glucose
    FROM
        {table_name}
    GROUP BY
        cigs_category
    """

    # Execute the query
    cursor.execute(query)

    # Fetch and print the results
    results = cursor.fetchall()
    for row in results:
        category, avg_age, avg_heart_rate, avg_sysBP, avg_glucose = row
        print(f"Category: {category}")
        print(f"Average Age: {avg_age:.2f}")
        print(f"Average Heart Rate: {avg_heart_rate:.2f}")
        print(f"Average Systolic BP: {avg_sysBP:.2f}")
        print(f"Average Glucose: {avg_glucose:.2f}")
        print()

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection to the database
    conn.close()
#cannot say anything on this
```
